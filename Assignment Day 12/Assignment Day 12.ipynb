```python
import pandas as pd
from scipy.stats import chi2_contingency,mannwhitneyu
```

```python
dataset=pd.read_csv("C:/Users/Chaitanya/LetsUpgrade-AI-ML/Assignment Day 12/general_data.csv")
dataset.head()
```

```python
dataset=dataset.drop_duplicates()
dataset=dataset.dropna()
dataset
```

```python
dataset["Attrition"].replace(to_replace=("No","Yes"),value=(0,1),inplace=True)
dataset
```

```python
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
dataset.Attrition = enc.fit_transform(dataset.Attrition)
dataset['BusinessTravel'] = enc.fit_transform(dataset['BusinessTravel'])
dataset['Department'] = enc.fit_transform(dataset['Department'])
dataset['EducationField'] = enc.fit_transform(dataset['EducationField'])
dataset['Gender'] = enc.fit_transform(dataset['Gender'])
dataset['JobRole'] = enc.fit_transform(dataset['JobRole'])
dataset['MaritalStatus'] = enc.fit_transform(dataset['MaritalStatus'])
dataset
```

```python
dataset_left=dataset[dataset['Attrition']== 1]
dataset_stayed=dataset[dataset['Attrition']== 0]
```

# Mann Whitney Test
1.

H0: There is no significant difference between the Ages of employees who left and employees who stayed.

H1: There is significant difference between the Ages of employees who left and employees who stayed.

```python
from scipy.stats import mannwhitneyu
stats, p = mannwhitneyu(dataset_left.Age, dataset_stayed.Age)
print(stats,p)
if p<0.05:
    print("H1 Accepted: There is significant difference between the Ages of employees who left and employees who stayed.")
else:
    print("H0 Accepted: There is no significant difference between the Ages of employees who left and employees who stayed.")
```

2.

H0: There is no significant difference between the Monthly Income of employees who left and employees who stayed.

H1: There is significant difference between the Monthly Income of employees who left and employees who stayed.

```python
from scipy.stats import mannwhitneyu
stats, p = mannwhitneyu(dataset_left.MonthlyIncome, dataset_stayed.MonthlyIncome)
print(stats,p)
if p<0.05:
    print("H1 Accepted: There is significant difference between the Monthly Income of employees who left and employees who stayed.")
else:
    print("H0 Accepted: There is no significant difference between the Monthly Income of employees who left and employees who stayed.")
```

# Chi Square Test
1.

H0: There is no relationship between Attrition and Marital Status.

H1: There is relationship between Attrition and Marital Status.

```python
chitable = pd.crosstab(dataset.Attrition,dataset.MaritalStatus)
chitable
```

```python
stats,p,dof,expected = chi2_contingency(chitable)
print(stats,p)
if p<0.05:
    print("H1 Accepted: There is relationship between Attrition and Marital Status.")
else:
    print("H0 Accepted: There is no relationship between Attrition and Marital Status.")
```

2.

H0: There is no relationship between Attrition and Gender.

H1: There is relationship between Attrition and Gender.

```python
chitable = pd.crosstab(dataset.Attrition,dataset.Gender)
chitable
```

```python
stats,p,dof,expected = chi2_contingency(chitable)
print(stats,p)
if p<0.05:
    print("H1 Accepted: There is relationship between Attrition and Gender.")
else:
    print("H0 Accepted: There is no relationship between Attrition and Gender.")
```
